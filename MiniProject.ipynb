```python
import tkinter as tk
from tkinter import simpledialog
import datetime

class Application(tk.Frame):
    def __init__(self, master=None):
        super().__init__(master)
        self.master = master
        self.grid()
        self.create_widgets()

    def create_widgets(self):
        self.master.title('Mini Project')
        
        self.student_info = tk.Label(self, text="Name: YOUR NAME, ID: YOUR ID")
        self.student_info.grid(row=0, column=0, columnspan=2)
        
        self.current_time = tk.Label(self, text=f"Current date and time: {datetime.datetime.now()}")
        self.current_time.grid(row=1, column=0, columnspan=2)

        self.entry1 = tk.Entry(self)
        self.entry2 = tk.Entry(self)
        self.entry3 = tk.Entry(self)
        self.entry4 = tk.Entry(self)

        self.entry1.grid(row=2, column=1)
        self.entry2.grid(row=3, column=1)
        self.entry3.grid(row=4, column=1)
        self.entry4.grid(row=5, column=1)

        self.text_widget = tk.Text(self)
        self.text_widget.grid(row=6, column=0, columnspan=2)
        
        self.add_button = tk.Button(self, text="ADD", command=self.add_to_file)
        self.add_button.grid(row=7, column=0)

        self.display_button = tk.Button(self, text="DISPLAY", command=self.display_from_file)
        self.display_button.grid(row=7, column=1)

        self.search_button = tk.Button(self, text="SEARCH", command=self.search)
        self.search_button.grid(row=8, column=0)

        self.clear_button = tk.Button(self, text="CLEAR", command=self.clear)
        self.clear_button.grid(row=8, column=1)

    def add_to_file(self):
        with open('data.txt', 'a') as f:
            f.write(self.entry1.get() + '\n')
            f.write(self.entry2.get() + '\n')
            f.write(self.entry3.get() + '\n')
            f.write(self.entry4.get() + '\n')
        self.entry1.delete(0, 'end')
        self.entry2.delete(0, 'end')
        self.entry3.delete(0, 'end')
        self.entry4.delete(0, 'end')

    def display_from_file(self):
        with open('data.txt', 'r') as f:
            self.text_widget.insert('end', f.read())

    def search(self):
        search_term = simpledialog.askstring("Search", "Enter search term:")
        with open('data.txt', 'r') as f:
            lines = f.readlines()
            matches = [line for line in lines if search_term in line]
            self.text_widget.insert('end', 'Matches:\n' + ''.join(matches))

    def clear(self):
        self.entry1.delete(0, 'end')
        self.entry2.delete(0, 'end')
        self.entry3.delete(0, 'end')
        self.entry4.delete(0, 'end')
        self.text_widget.delete('1.0', 'end')

root = tk.Tk()
app = Application(master=root)
app.mainloop()

```
